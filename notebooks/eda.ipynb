# Vaccine Sentiment — Exploratory Data Analysis
Run `generate_dataset.py` and `train.py` before executing this notebook.

```python
import pandas as pd
import matplotlib.pyplot as plt
import sys, os
sys.path.insert(0, '..')
from preprocess import clean_text

df = pd.read_csv('../dataset.csv')
print(df.shape)
df.head()
```

```python
# Label distribution
df['label'].value_counts().plot(kind='bar', color=['#2ecc71','#e74c3c','#3498db'])
plt.title('Sentiment Distribution')
plt.show()
```

```python
# Tweet length distribution
df['tweet_len'] = df['tweet'].apply(len)
df.groupby('label')['tweet_len'].hist(alpha=0.7, bins=20)
plt.xlabel('Tweet Length (characters)')
plt.title('Tweet Length by Sentiment')
plt.legend(df['label'].unique())
plt.show()
```

```python
# Preview preprocessed text
df['clean'] = df['tweet'].apply(clean_text)
df[['tweet','clean','label']].head(10)
```
